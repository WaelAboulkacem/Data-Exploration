# Supply Chain KPI Definition

To assess supply chain health and planning risks, KPIs are grouped into four main dimensions:
- Inventory & Demand Risk
- Lead Time & Planning Risk
- Cost & Efficiency
- Quality & Reliability

These KPI categories guide the selection and computation of metrics in the following steps.


### Inventory & Demand Risk KPIs

The following KPIs are defined to evaluate inventory adequacy and stock-out risk:

- **Stock Coverage Ratio:** Measures how well current stock levels can satisfy observed demand.
- **Stock-Out Flag:** Identifies products with zero stock, indicating immediate availability risk.
- **Demand Pressure Index:** Captures demand intensity relative to available inventory.

These metrics help highlight products exposed to inventory shortages or imbalance.


### Lead Time & Planning Risk KPIs

The following KPIs are defined to assess replenishment speed and planning risk:

- **Average Total Lead Time:** Captures the overall time required to replenish products across the supply chain.
- **Manufacturing Lead Time Share:** Measures the proportion of total lead time attributable to manufacturing activities.
- **Long Lead Time Flag:** Identifies products with above-median lead times, indicating potential planning risk.

These KPIs help highlight delays and bottlenecks within the supply chain.


### Cost & Efficiency KPIs

The following KPIs are defined to evaluate cost structure and operational efficiency:

- **Cost per Unit Sold:** Measures overall supply chain cost efficiency relative to sales volume.
- **Shipping Cost Intensity:** Assesses logistics cost pressure compared to generated revenue.
- **Manufacturing Cost Intensity:** Evaluates the proportion of revenue consumed by manufacturing costs.

These metrics help identify cost drivers and efficiency improvement opportunities.


### Quality & Reliability KPIs

The following KPIs are defined to assess quality performance and operational reliability:

- **Defect Rate:** Measures the proportion of defective products, indicating quality risk.
- **Inspection Failure Rate:** Captures the frequency of failed inspections.
- **Quality Cost Exposure:** Estimates the potential financial impact of quality issues.

These KPIs help identify reliability risks and quality-related cost drivers.
